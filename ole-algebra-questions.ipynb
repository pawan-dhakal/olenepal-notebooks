```python
np.random.seed(10) #for testing, use same random seed, if needed
def gen_equations_type1(level=1,num=1):
    rand_letter_list = np.array([]);
    for i in range(num):
        rand_letter = random.choice(string.ascii_letters[:26]) #small letters only
        rand_letter_list = np.append(rand_letter_list,rand_letter)
    
    if level == 1:
        min_no = 1
        max_no = 10
        operation_str = ' + '
    elif level == 2:
        min_no = 1
        max_no = 10
        operation_str = ' - '
    elif level == 3:
        min_no = 11
        max_no = 50
        operation_str = ' + '
    elif level == 4:
        min_no = 11
        max_no = 50
        operation_str = ' - '
        
    rand_no1 = np.random.randint(min_no,max_no,size=num)
    rand_no2 = np.random.randint(min_no,max_no,size=num)
    
    q_string_list = np.array([]);
    ans_string_list = np.array([]);
    
    opt1 = rand_no1; #first two options are the two numbers in the question
    opt2 = rand_no2;
    opt3 = opt1+opt2; #for both + and -, this gives wrong answer
    opt1 = np.array([str(r) for r in opt1]); #convert to string 
    opt2 = np.array([str(r) for r in opt2]);
    opt3 = np.array([str(r) for r in opt3]);

    for i in range(num):
        #compare rand no and make sure that answer is not negative
        #take the bigger number of the two as the final number (for addition; reverse for subtraction)
        no1 = min(rand_no1[i],rand_no2[i])
        no2 = max(rand_no1[i],rand_no2[i])
        var = rand_letter_list[i]
        if operation_str == ' + ':
            q_string = str(no1) + operation_str + str(var) + ' = ' + str(no2)
            ans_string = str(no2-no1) #find correct answer (big - small = val of var)
        if operation_str == ' - ':
            q_string = str(no2) + operation_str + str(var) + ' = ' + str(no1)
            ans_string = str(no2-no1) #check
        q_string_list = np.append(q_string_list,q_string)
        ans_string_list = np.append(ans_string_list,ans_string)
        #if any opt == ans, change opt
        if opt1[i]==ans_string:
            opt1[i] = str(int(opt1[i])+1)
        if opt2[i]==ans_string:
            opt2[i] = str(int(opt2[i])+1)
        if opt3[i]==ans_string:
            opt3[i] = str(int(opt3[i])+1)
    #remove questions with answers = 0
    non_zero_mask = ans_string_list !='0'
    ans_string_list = ans_string_list[non_zero_mask]
    q_string_list = q_string_list[non_zero_mask]
    opt1 = opt1[non_zero_mask]
    opt2 = opt2[non_zero_mask]
    opt3 = opt3[non_zero_mask]
    return q_string_list, ans_string_list, opt1, opt2, opt3 
```

```python
num = 10
print('remove questions with answers=0')
for level in [1,2,3,4]:
    print('\n Level: %d \n'%level)
    questions, answers, opt1, opt2, opt3 = gen_equations_type1(level=level,num=num)
    print(len(answers),len(opt1))
    for i in range(len(questions)):
        print(opt1[i]==answers[i],opt2[i]==answers[i],opt3[i]==answers[i])
        print('ques no (%d): '%(i+1),questions[i],', ans = ',answers[i], ' options : [%s, %s, %s]'%(opt1[i],opt2[i],opt3[i]))
```

```python
np.random.seed(10) #for testing, use same random seed, if needed
def gen_equations_type2(level=1,num=1,name_list=['Tenzin','Hari','Sita','Rabi','Regan','Anjana','Deepa'],obj_list=['khatas','pencils','notebooks','rulers','books']):
    #get numbers and results from type1 
    questions, answers = gen_equations_type1(level=level,num=num)
    
    obj1 = random.choice(obj_list)
    name1 = random.choice(name_list)
    #name2 = name1
    name2 = name1;
    while name2 == name1: #two names must be different for the question
        name2 = random.choice(name_list)
        if name2 != name1:
            break
    q_words_list = np.array([]);
    a_words_list = np.array([]);
    for i in range(len(questions)):
        q = questions[i].split(); ans = answers[i].split();
        num1 = q[0]; num2 = q[-1]; oper = q[1];
        sent1 = name1 + ' had %s %s yesterday. '%(num1, obj1)
        if oper == '+':    
            sent2 = '%s got some more %s from %s. '%(name1, obj1,name2) 
            sent4 = 'How many %s did %s get from %s? '%(obj1, name1, name2)
        else:
            sent2 = '%s gave some %s to %s. '%(name1, obj1,name2) 
            sent4 = 'How many %s did %s give to %s? '%(obj1, name1, name2)
        sent3 = 'Now %s has %s %s in total. '%(name1, num2,obj1)
        q_words = sent1 + sent2 + sent3 + sent4
        q_words_list = np.append(q_words_list,q_words)
        a_words_list = np.append(a_words_list,ans)
    return q_words_list, a_words_list
```
