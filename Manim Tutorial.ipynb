```python
%%manim -qm solveLinearEquation

class solveLinearEquation(Scene):
    def construct(self):
        eq1 = MathTex("{{ x }} + {{ 2 }} = {{ 5 }}")
        eq2 = MathTex("{{ x }} + {{ 2 }} - {{ 2 }} = {{ 5 }} - {{ 2 }}")
        eq3 = MathTex("{{ x }} + 0 = {{ 3 }}")
        eq4 = MathTex("{{ x }} = {{ 3 }}")
        self.add(eq1)
        self.wait()
        self.play(TransformMatchingTex(eq1,eq2))
        self.wait()
        self.play(TransformMatchingTex(eq2,eq3))
        self.wait()
        self.play(TransformMatchingTex(eq3,eq4))
        self.wait()
        
```

```python
%%manim -qm solveLinearEquation2

class solveLinearEquation2(Scene):
    def construct(self):
        eq1 = MathTex("2 {{ x }} + {{ 2 }} = {{ 6 }}")
        eq2 = MathTex("2 {{ x }} = {{ 6 }} - {{ 2 }}")
        eq3 = MathTex("2 {{ x }} = {{ 4 }} ")
        eq4 = MathTex(r"x =  \frac{4}{2} ")
        eq5 = MathTex("{{ x }} = 2")
        self.add(eq1)
        self.wait()
        self.play(TransformMatchingTex(eq1,eq2))
        self.wait()
        self.play(TransformMatchingTex(eq2,eq3))
        self.wait()
        self.play(TransformMatchingShapes(eq3,eq4))
        self.wait()
        self.play(TransformMatchingTex(eq4,eq5))
        self.wait()
        
```

```python
MathTex(
            r"d  = ",
            r"2 \textrm{km} = ",
            r"2 \times 1000 \textrm{m} = ",
            r"2000 \textrm{m}"
        )
```

```python
%%manim -qm FormulaEmphasis

class FormulaEmphasis(Scene):
    def construct(self):
        distance = MathTex(
            r"d  = 2 \textrm{km}",
            r"= ",
            r"2 \times 1000 \textrm{m} ",
            r"= ",
            r"2000 \textrm{m}"
        )
        self.play(Write(distance))
        box1 = SurroundingRectangle(distance[0], buff=0.1)
        box2 = SurroundingRectangle(distance[2], buff=0.1)
        box3 = SurroundingRectangle(distance[4], buff=0.1)
        self.play(Create(box1))
        self.wait()
        self.play(Transform(box1, box2))
        self.wait()
        self.play(Transform(box1, box3))
        self.wait()
```

# Logo

```python
%%manim -qm ManimCELogo

class ManimCELogo(Scene):
    def construct(self):
        self.camera.background_color = "#ece6e2"
        logo_green = "#87c2a5"
        logo_blue = "#525893"
        logo_red = "#e07a5f"
        logo_black = "#343434"
        ds_m = MathTex(r"\mathbb{M}", fill_color=logo_black).scale(7)
        ds_m.shift(2.25 * LEFT + 1.5 * UP)
        circle = Circle(color=logo_green, fill_opacity=1).shift(LEFT)
        square = Square(color=logo_blue, fill_opacity=1).shift(UP)
        triangle = Triangle(color=logo_red, fill_opacity=1).shift(RIGHT)
        logo = VGroup(triangle, square, circle, ds_m)  # order matters
        logo.move_to(ORIGIN)
        self.add(logo)
        self.play(logo.animate.shift(UP).scale(2))
```

```python
%%manim -qm OLENepalLogo

class OLENepalLogo(Scene):
    def construct(self):
        
        logo_yellow = "#f8c022"
        logo_red = "#db282e"
        logo_green = "#00a651"
        logo_blue = "#0095da"
        logo_black = "#3a3a3a"
        self.camera.background_color = logo_black
        text_a = Text("").scale(2)
        #self.add(text_a)
        #creating arms
        # yellow
        arc_y= ArcBetweenPoints(start= LEFT, end= RIGHT, angle=PI,stroke_color=logo_yellow, stroke_width=20)
        line_y= Line(start=[0., -0.5, 0.], end=[0., -1., 0.],stroke_color=logo_yellow, stroke_width=20)
        dot_y = Dot(point=ORIGIN, radius=0.5, stroke_width=1, fill_opacity=1.0, color=logo_yellow)
        arm_yellow = VGroup(arc_y,line_y,dot_y)
        arm_yellow.move_to(ORIGIN).shift(1.2*LEFT).shift(1.2*UP).rotate(PI/4)
        #self.add(arm_yellow)
        #red
        arc_r= ArcBetweenPoints(start= LEFT, end= RIGHT, angle=PI,stroke_color=logo_red, stroke_width=20)
        line_r= Line(start=[0., -0.5, 0.], end=[0., -1., 0.],stroke_color=logo_red, stroke_width=20)
        dot_r = Dot(point=ORIGIN, radius=0.5, stroke_width=1, fill_opacity=1.0, color=logo_red)
        arm_red = VGroup(arc_r,line_r,dot_r)
        arm_red.move_to(ORIGIN).shift(1.2*RIGHT).shift(1.2*UP).rotate(-PI/4)
        #self.add(arm_red)
        #blue 
        arc_b= ArcBetweenPoints(start= LEFT, end= RIGHT, angle=PI,stroke_color=logo_blue, stroke_width=20)
        line_b= Line(start=[0., -0.5, 0.], end=[0., -1., 0.],stroke_color=logo_blue, stroke_width=20)
        dot_b = Dot(point=ORIGIN, radius=0.5, stroke_width=1, fill_opacity=1.0, color=logo_blue)
        arm_blue = VGroup(arc_b,line_b,dot_b)
        arm_blue.move_to(ORIGIN).flip(axis=[1., 0., 0.]).shift(1.2*DOWN).shift(1.2*RIGHT).rotate(PI/4) #flip about x axis
        #self.add(arm_blue)
        #green 
        arc_g= ArcBetweenPoints(start= LEFT, end= RIGHT, angle=PI,stroke_color=logo_green, stroke_width=20)
        line_g= Line(start=[0., -0.5, 0.], end=[0., -1., 0.],stroke_color=logo_green, stroke_width=20)
        dot_g = Dot(point=ORIGIN, radius=0.5, stroke_width=1, fill_opacity=1.0, color=logo_green)
        arm_green = VGroup(arc_g,line_g,dot_g)
        arm_green.move_to(ORIGIN).flip(axis=[1., 0., 0.]).shift(1.2*LEFT).shift(1.2*DOWN).rotate(-PI/4)
        #self.add(arm_green)
        
        logo = VGroup(text_a,  arm_red, arm_yellow, arm_green, arm_blue)  # order matters
        logo.move_to(ORIGIN)
        self.add(logo)

        #logo.shift(5*LEFT).scale(.5)
        self.play(logo.animate.shift(5*LEFT).scale(.5))
        self.wait()

        logoText1 = Text("  -").scale(1.5).shift(0.5*UP)
        self.play(Create(logoText1))
        self.wait()

        logoText2 = Text("OLE NEPAL").scale(1).shift(0.5*DOWN).shift(1.4*LEFT)
        self.play(Create(logoText2))
        self.wait(1)

        all_logo = VGroup(logo,logoText1,logoText2)
        self.play(all_logo.animate.move_to(ORIGIN).scale(.7))
        self.wait()
        self.play(all_logo.animate.scale(1.2))
        self.wait()
```
